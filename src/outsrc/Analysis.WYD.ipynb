```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
wyd = pd.io.stata.read_stata("WorldIncomeDistribution1988-2005.dta")
```

```python
wyd = wyd.loc[wyd.year == 2005,:]
```

```python
plotme = wyd.loc[wyd.group == 1.0, ("country","Dinc","pop")]
```

```python
plotme.plot.scatter(x="pop",y="Dinc", c='c', s=50)
```
