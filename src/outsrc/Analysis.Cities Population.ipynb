```python
plot50 = statistics2017.sort_values(by="1000hb_city", ascending=False).head(50)
```

```python
#Per 1000 hab 
group_names= plot50.country
group_size= plot50["1000hb_city"]
subgroup_names= plot50.city
subgroup_size=plot50["1000hb_city"]
 
# Create colors
cmap = plt.get_cmap("tab20c")
outer_colors = cmap(np.arange(21)*1)
inner_colors = cmap(np.arange(21)*1)
 
# First Ring (outside)
fig, ax = plt.subplots(figsize=(20,20))
mypie, _ = ax.pie(group_size, radius=1.3, labels=group_names, colors=outer_colors)
plt.setp( mypie, width=0.3, edgecolor='white')
 
# Second Ring (Inside)
mypie2, _ = ax.pie(subgroup_size, radius=1.3-0.3, labels=subgroup_names, labeldistance=0.8, colors=inner_colors)
plt.setp( mypie2, width=0.4, edgecolor='white')
plt.margins(0,0)

plt.show()
```
