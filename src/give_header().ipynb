```python
def read_file(in_path):
    with open(in_path, encoding="Latin-1") as f:
        read_content = f.read()
        return read_content
        
        
def write_file(in_path, out_path, string):
        with open(out_path, "w", encoding="Latin-1") as f_new:
            read_content = read_file(in_path)
            f_new.write(string + "\n")
            f_new.write(read_content)
```

```python
def apply_list(list, base_in_path, base_out_path, string):
    for file in list:
        in_path = base_in_path + file
        out_path = base_out_path + file
        read_file(in_path)
        write_file(in_path, out_path, string)
        
#Full index:
apply_list(OA_list,"./../data/full/headerless/OA_status_raw/" ,"./../data/full/OA_status_raw/", "doi,is_oa,data_standard,best_oa_location_url,best_oa_location_host_type,best_oa_location_version,best_oa_location_licence,best_oa_location_evidence,title,journal_issns,journal_name,journal_is_oa,publisher,updated")
#Test run:
#apply_list(OA_list,"./../data/sample/OA_status//" ,"./../data/sample/test/", "doi,is_oa,data_standard,best_oa_location_url,best_oa_location_host_type,best_oa_location_version,best_oa_location_licence,best_oa_location_evidence,title,journal_issns,journal_name,journal_is_oa,publisher,updated")
```
