```python
import pandas as pd
```

```python
OA_list = ["oa_status00","oa_status01","oa_status02","oa_status03","oa_status04","oa_status05","oa_status06","oa_status07","oa_status08","oa_status09","oa_status10","oa_status11","oa_status12","oa_status13","oa_status14","oa_status15","oa_status16","oa_status17","oa_status18","oa_status19"]
```

```python
def read_doi_oa(in_path):
    #import from (path + list), 'doi' and 'is_oa' columns.
    with open(in_path, encoding="Latin-1") as f:
        data = pd.read_csv(f, usecols=["doi", "is_oa"], index_col=0, header=0)
        #After column rename, replace same column with boolean values.
        data = data.replace({"is_oa":{"f":False,"t":True}})
        
        return data

def read_sample(in_path):
    with open(in_path, encoding="Latin-1") as f:
        data = pd.read_csv(f, usecols=["doi", "is_oa"], index_col=0 ,header=0 )
        
        #Extract 25 line sample.
        return data.sample(25)

def apply_func(file_list, func,base_out_path,base_in_path):
    for file in file_list:
        in_path = base_in_path + file
        out_path = base_out_path + file
        
        data = func(in_path)
        
        #Export to new directory, full, clean, dataframe.
        data.to_csv(out_path)
```

```python
apply_func(OA_list, read_doi_oa, "./../data/full/OA_status/", "./../data/full/OA_status_raw/")
```

```python
apply_func(OA_list, read_sample, "./../data/sample/OA_status/", "./../data/full/OA_status/")
```
