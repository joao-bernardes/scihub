```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from collections import Counter
import seaborn as sns
```

## Import DataFrame

```python
statistics2017 = pd.read_csv("./../../data/sample/2017.statistics.sample.with_missing_cities_continents.csv")
```

```python
statistics2017 = statistics2017.loc[statistics2017.lon >0,:]
statistics2017 = statistics2017.loc[statistics2017.continent == "Oceania",:]
statistics2017
```

```python
def show_cart():
    fig = plt.figure(figsize=(20,20))
    ax = plt.axes(projection=ccrs.PlateCarree())
    ax.coastlines( color='black', linewidth=1)
    ax.add_feature(cfeature.BORDERS, linestyle=':', linewidth=0.5 )

def scatter_points(lat, lon):
    c = Counter(zip(lon,lat))
    area = [0.1 * c[(lonlon,latlat)]*100 for lonlon,latlat in zip(lon,lat)]
    plt.scatter(lon, lat, color="darkred" ,marker='.', s=area)

def plotmap(df, title):
    for i in df.drop_duplicates():
        lat=df.lat
        lon=df.lon
    show_cart()
    scatter_points(lat, lon)
    #plt.title(title)
    #plt.savefig('./../../charts and figures/'+ "Analysis.2017.Downloads" + b + ".pdf")
    plt.show()
```

```python
plotmap(statistics2017," Article Consumption Estimate ")
    
```
