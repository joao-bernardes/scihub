```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
internet_access = pd.read_csv("./API_IT.NET.USER.ZS_DS2_en_csv_v2.csv", usecols=("Country Name", "2016"))
```

```python
internet_access.head()
```

```python
internet_access["bin"] = pd.cut(internet_access["2016"] ,3, labels=["low", "medium", "high"])
```

```python
internet_access.head()
```

```python
internet_access = internet_access.dropna()
```

```python
internet_access.loc[internet_access.bin == "high",:].sort_values(by="2016", ascending=False).plot.bar(figsize=(40,15), fontsize=20, x="Country Name" )
```

```python
internet_access.loc[internet_access.bin == "medium",:].sort_values(by="2016", ascending=False).plot.bar(figsize=(40,15), fontsize=20,x="Country Name")
```
