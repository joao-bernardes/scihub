```python
import pandas as pd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
statistics2017 = pd.read_csv("./../../data/sample/2017.statistics.sample.with_missing_cities_continents.csv")
statistics2017.info()
```

```python
table1 = statistics2017.loc[statistics2017.continent == "Africa",("country","DOI")].groupby('country').count().sort_values(by='DOI',ascending = False).head(10)
table1 = table1.rename({"DOI": "Downloads"}, axis = "columns")
table1
```

```python
print (table1.to_latex())
```

```python
table2 = statistics2017.loc[statistics2017.continent == "Africa",("country","city","DOI")].groupby(['country','city']).count().sort_values(by='DOI',ascending = False).head(10)
table2 = table2.rename({"DOI": "Downloads"}, axis = "columns")
table2
```

```python
print (table2.to_latex())
```

```python
table3 = statistics2017.loc[statistics2017.continent == "Africa",("country","DOI")].groupby('country').count().sort_values(by='DOI').head(10)
table3 = table3.rename({"DOI": "Downloads"}, axis = "columns")
table3
```
